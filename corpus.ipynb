```python
def get_date(b_soup):
    article_date = b_soup.find('label', class_='css-hn0k3p')
    try:
        article_date = article_date.get_text(' ', strip=True)
    except:
        article_date = b_soup.find('div', class_='css-1kkxezg')
        if article_date is None:
            article_date = b_soup.find('time', class_='content__dateline-wpd js-wpd')
            article_date = article_date.get_text(' ', strip=True)
        else:
            article_date = article_date.get_text(' ', strip=True)

    return article_date


def get_topics(b_soup):

    topics = b_soup.find('meta', property='article:tag')
    return topics['content'].replace(',', ', ')


def get_texts(l, dd, article_title, article_link):

    html = urlopen(article_link, context=ctx).read()
    soup = BeautifulSoup(html, 'html.parser')

    d = dict()

    # Retrieve the full text of the article
    full_text = soup('div', class_='article-body-commercial-selector')   
    if not full_text:
        full_text = soup('div',
        class_='content__article-body from-content-api js-article__body')

    # Remove text after bullet
    bad_tag = soup.find('span', {'data-dcr-style' : 'bullet'})

    # Retrieve the date of the article
    date = get_date(soup)

    for tag in full_text:
        
        # Retrieve the topics of the article
        topics = get_topics(soup)

        # Retrive the initial letter, if exists
        try:
            letter = tag.span.get_text(strip=True)
            if len(letter) == 1:
                text = f"{letter}{tag.get_text(' ', strip=True)[2:]}"
            else:
                text = tag.get_text(' ', strip=True)
        except:
            text = tag.get_text(' ', strip=True)

        if bad_tag:
            for sibling in bad_tag.find_parent().next_siblings:
                try:
                    text = text.replace(sibling.get_text(' ', strip=True), ' ')
                except:
                    text = text.replace(sibling.string, ' ')
            for sibling in bad_tag.next_siblings:
                if sibling.string: text = text.replace(sibling.string, ' ')
        
        # Remove ad, author names, etc.
        for line in bad_lines[:3]:
            if line in text:
                text = text.replace(f'{name} {line}', ' ')
        for line in bad_lines[3:]:
            if line in text:
                text = text.replace(line, ' ')
        
        # Remove photo descriptions
        text = re.sub(r'Photograph: (.*?(Getty Images|PA|Reuters|AP|EPA|Shutterstock|The Guardian|Alamy|BBC|Observer|Rex|REX|Channel 4|ITV)\s)',
                      '', text)        
    
        
        d = {'Author': name, 'Title': article_title, 'Date': date,
        'Article URL': article_link, 'Full text': text, 'Topics': topics}
        l.append(d)
        dd['Texts'] = l

        # Write the data to different JSON files
        with open(f"texts_{(i+1):02d}_{name.replace(' ', '_')}.json", 'w') as fh3:
            json.dump(dd, fh3, indent=4)


def get_files(author):
    
    l, dd = list(), dict()

    with open(f"articles_{(i+1):02d}_{author.replace(' ', '_')}.json", 'r') as fh2:
        js2 = json.load(fh2)
    for j in range(100):
        title = js2['Articles'][j]['Title']
        link = js2['Articles'][j]['Article URL']
        get_texts(l, dd, title, link)



with open('columnists.json', 'r') as fh:
    js = json.load(fh)
for i in range(0, 1):
    name = js['Columnists'][i]['Name']
    get_files(name)

fh.close()
```
